### Get Formatted ISBN

```python
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
import time

import os
chromedriver = "/Applications/chromedriver"      # path to the chromedriver executable
os.environ["webdriver.chrome.driver"] = chromedriver
```

```python
def get_isbn(row):
    try: 
        if row == "0":
            return "N/A"
        else:
            driver.find_element_by_name("isbn_in").send_keys(row)
            driver.find_element_by_xpath("/html/body/form/div/table/tbody/tr[3]/th/input[3]").click()
            isbn = driver.find_element_by_name("isbn_in").get_attribute("value")
            driver.find_element_by_name("isbn_in").clear()
            return isbn
    except:
        try:
            alert = driver.switch_to_alert()          # if a pop up alert happened, if so, click it
            alert.accept()
            return "N/A"
        except: 
            return "N/A"
```

```python
driver = webdriver.Chrome(chromedriver)

driver.get("http://pcn.loc.gov/isbncnvt.html")    # Library of Congress website that formats ISBNs properly 
driver.find_element_by_name("hyphenate").click()  # check the hyphenate box

inventory["isbn-format"] = inventory["isbn"].apply(get_isbn) 
```
